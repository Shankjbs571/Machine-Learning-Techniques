```python
# prompt: how to access a csv file from drive link and make it a dataframe

import pandas as pd

df = pd.read_csv('V5.csv')
df.head()
```

```python
df.shape
```

```python
# Replace "?" with NaN
df.replace("?", float("NaN"), inplace=True)
df.head()
```

```python
nan_counts = df.isna().sum()
print(nan_counts)

# Count the total number of NaN values in the entire DataFrame
total_nan_count = nan_counts.sum()

# Display the total count
print("Total NaN count:", total_nan_count)
```

```python

# Get the list of column names
column_names = df.columns

# Iterate over each column and count the number of missing values
for column in column_names:
    nan_count = df[column].isna().sum()
    if nan_count > 0:
        print(f"Feature '{column}' has {nan_count} missing values")
```
