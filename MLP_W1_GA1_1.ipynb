```python
# prompt: how to access a csv file from drive link and make it a dataframe

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/V5.csv')
df.head()
```

```python

# Get the list of column names
column_names = df.columns

# Iterate over each column and count the number of missing values
for column in column_names:
    nan_count = df[column].isna().sum()
    if nan_count > 0:
        print(f"Feature '{column}' has {nan_count} missing values")
```

```python
# prompt: drop all rows having missing values greater than 2

df = df.dropna(thresh=2)
df.shape
```

```python
# prompt: select all even indexed rows of df

even_indexed_rows = df[df.index % 2 == 0]
even_indexed_rows.iloc[332,3]
```

```python
# Select all the odd indexed rows from the given dataset. What is the value in the 3rd indexed column of the 332th indexed row in the selected dataframe.

odd_indexed_rows = df[df.index % 2 != 0]
odd_indexed_rows.iloc[100, 5]
```

```python
# prompt:  Determine the number of samples (rows) in the dataset that belong to the six most recent years, based on the provided data.


# Extract the year from the 'Date' column
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Get the six most recent years
recent_years = sorted(df['Year'].unique())[-6:]
recent_years

```

```python
# Count the number of samples in the last six years
recent_samples_count = df[df['Year'].isin(recent_years)].shape[0]

# Printing the result
print(f"Number of samples in the last six years: {recent_samples_count}")
```

```python
#Calculate the number of samples(rows) that correspond to the month of August across all the years.?

df['Month'] = pd.to_datetime(df['Date']).dt.month

# Count the number of samples in August
august_samples_count = df[df['Month'] == 8].shape[0]

# Printing the result
print(f"Number of samples in August: {august_samples_count}")
```

```python
# According to the data which Locality have highest average Sale Price?

# Group the data by locality and calculate the average sale price
average_sale_price_by_locality = df.groupby('Locality')['Sale Price'].mean()
average_sale_price_by_locality.head()
```

```python

# Find the locality with the highest average sale price
locality_with_highest_average_sale_price = average_sale_price_by_locality.sort_values(ascending=False).index[0]

# Print the result
print(f"Locality with the highest average sale price: {locality_with_highest_average_sale_price}")
```
