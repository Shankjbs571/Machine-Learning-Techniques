```python
#@title (1) What is the average of the flight ticket price? Write your answer correct to two decimal places.
avg_price = df['Price'].mean()
avg_price.round(3)
```

```python
#@title (2)During which month did the highest number of flights occur? Months are represented by numerical codes, with January corresponding to 1, February to 2, and so forth.
print(df['Month'].unique())
month_wise_group = df.groupby('Month').size()
month_wise_group

```

```python
#@title (3) Is the average price of flight tickets higher on weekends (Saturday and Sunday) or on weekdays (Remaining 5 days)?

df['Weekend'] = df['WeekDay'].isin(['Saturday ', 'Sunday'])  # True for weekend, False for weekday
average_price_weekend = df[df['Weekend']]['Price'].mean()
average_price_weekday = df[~df['Weekend']]['Price'].mean()

if average_price_weekend > average_price_weekday:
    print("Average price of flight tickets is higher on weekends.")
else:
    print("Average price of flight tickets is higher on weekdays.")

```

```python
#@title (4)Two of the entries in the 'Additional_Info' column are 'No info' and 'No Info'. Replace all occurrences of 'No Info' with 'No info'. How many flights fall under airline 'IndiGo' and have 'No info' as additional information?

df['Additional_Info']=df['Additional_Info'].replace('No Info','No info')
result = df[(df['Airline']=='IndiGo') & (df['Additional_Info'] == 'No info')]
result.shape[0]
```
