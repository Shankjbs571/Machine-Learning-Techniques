```python
#@title (3) Is the average price of flight tickets higher on weekends (Saturday and Sunday) or on weekdays (Remaining 5 days)?

df['Weekend'] = df['WeekDay'].isin(['Saturday ', 'Sunday'])  # True for weekend, False for weekday
average_price_weekend = df[df['Weekend']]['Price'].mean()
average_price_weekday = df[~df['Weekend']]['Price'].mean()

if average_price_weekend > average_price_weekday:
    print("Average price of flight tickets is higher on weekends.")
else:
    print("Average price of flight tickets is higher on weekdays.")

```

```python
#@title (5) Convert the values of 'Duration' into seconds. Enter the average duration (in seconds) of a flight. Enter your answer correct to two decimal places.

def convert_to_seconds(duration):
    if 'h' in duration and 'm' in duration:
        hours, minutes = duration.split('h ')
        return int(hours) * 3600 + int(minutes.replace('m', '')) * 60
    elif 'h' in duration:
        return int(duration.split('h')[0]) * 3600
    elif 'm' in duration:
        return int(duration.replace('m', '')) * 60
    else:
        return 0  # Handle cases with no 'h' or 'm'

df['Duration_seconds'] = df['Duration'].apply(convert_to_seconds)
df.head()
df['Duration_seconds'].mean()
```

## Apply the following functions to the columns Dep_Time and Arrival_Time:

Transform the values in the 'dep_time' and 'arrival_time' columns to represent the hour component. For instance, if an entry is 10:05 June 13 or 10:05, the corresponding value should be 10.
Then convert the time into four categories as follows:
* 5 <= hour < 12 = Morning
* 12 <= hour < 17 = Afternoon
* 17 <= hour < 20 = Evening
* 20 <= hour < 5 = Night




```python

df['Dep_hour'] = df['Dep_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)
df['Arr_hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)

# Function to convert hour to time of day category
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 20:
        return 'Evening'
    elif 20 <= hour or hour < 5:
        return 'Night'

# Apply the function to the 'Hour' column to create a new column 'TimeOfDay'
df['TimeOfDep'] = df['Dep_hour'].apply(categorize_time)
df['TimeOfArr'] = df['Arr_hour'].apply(categorize_time)
df.drop(['Dep_hour','Arr_hour'],axis='columns')
```

```python
#@title (6) How many flights started in the Morning and arrived the destination at Evening?
flight_morning_to_evening = df[(df['TimeOfDep']=='Morning') & (df['TimeOfArr']=='Evening')]
flight_morning_to_evening.shape[0]
```

```python
#@title (7) Encode the values of column 'WeekDay' as follows: Weekends (Sunday, Saturday) = 1 , all remaining five days = 0 What is the most frequent (mode) WeekDay
def encode(day):
  if day in ['Sunday', 'Saturday']:
    return 1
  else:
    return 0

df['Encoded_WeekDay'] = df['WeekDay'].apply(encode)

count_1 = df['Encoded_WeekDay'].value_counts()
```

```python
count_1.loc[0]
if count_1.loc[0] > count_1.loc[1]:
  mode = 'WeekDays'
else:
  mode = 'Weekends'

print(f'The Most Frequent Mode is {mode}')
```
